# 서울시 대학 좌표 변경

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../../data/서울시_대학_주소.csv', encoding='euckr')
df
```

```python
# 카카오API를 사용하여 주소->좌표 변환
import requests, json

def get_location(address):
  url = 'https://dapi.kakao.com/v2/local/search/address.json?query=' + address
  # 'KaKaoAK '는 그대로 두시고 개인키만 지우고 입력해 주세요.
  # ex) KakaoAK 6af8d4826f0e56c54bc794fa8a294
  headers = {"Authorization": "KakaoAK "}
  api_json = json.loads(str(requests.get(url,headers=headers).text))
  address = api_json['documents'][0]['address']
  crd = {"lat": str(address['y']), "lng": str(address['x'])}
  address_name = address['address_name']

  return crd

crd = get_location("제주 애월읍")
print(crd)
```

```python
addr_list = []
addr_list.append(df['주소'])
print(addr_list)
```

```python
crd = []
for i in addr_list :
    for n in range(64) : # crd 리스트 안에 좌표값 넣기
        print(n, '값을 넣었습니다.')
        try :
            crd.append(get_location(i[n]))
        except :
            crd.append({'lat' : '0', 'lng' : '0'})
```

```python
# 리스트에 lng, lat 값을 넣고 df에 추가
lng_list = []
lat_list = []
for i in range(64) :
    lng_list.append(crd[i]['lng'])
    lat_list.append(crd[i]['lat'])
    
df['lng'] = lng_list
df['lat'] = lat_list

df
```

```python
df.to_csv("../../data/대학교_좌표추가.csv", mode='w', encoding='euckr')
```
