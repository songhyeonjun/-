# 파이썬 라이브러리를 이용한 방법

```python
from geopy.geocoders import Nominatim
```

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('서울특별시_전동킥보드_견인_현황_20220425.csv', encoding='euckr')
```

```python
df
```

```python
def geocoding(address):
    geolocoder = Nominatim(user_agent = 'South Korea', timeout=None)
    geo = geolocoder.geocode(str(address))
    crd = {"lat": str(geo.latitude), "lng": str(geo.longitude)}

    return crd
```

```python
a = []
a.append(df['주소'])
print(a)
```

```python
for i in a :
    for n in range(100) :
        print(i[n])
        try :
            crd = geocoding(i[n])
        except :
            crd = '값을 넣을 수 없어요'
        print(crd)
#         print(crd['lat'])
#         print(crd['lng'])
```

# 카카오 API 이용

```python
# 카카오API를 사용하여 주소->좌표 변환
import requests, json

def get_location(address):
  url = 'https://dapi.kakao.com/v2/local/search/address.json?query=' + address
  # 'KaKaoAK '는 그대로 두시고 개인키만 지우고 입력해 주세요.
  # ex) KakaoAK 6af8d4826f0e56c54bc794fa8a294
  headers = {"Authorization": "KakaoAK "}
  api_json = json.loads(str(requests.get(url,headers=headers).text))
  address = api_json['documents'][0]['address']
  crd = {"lat": str(address['y']), "lng": str(address['x'])}
  address_name = address['address_name']

  return crd

crd = get_location("제주 애월읍")
print(crd)
```

## 전동킥보드 견인 지역 좌표변경

```python
crd = []
for i in a :
    for n in range(43128) : # crd 리스트 안에 좌표값 넣기
        print(n, '값을 넣었습니다.')
        try :
            crd.append(get_location(i[n]))
        except :
            crd.append({'lat' : '0', 'lng' : '0'})
```
